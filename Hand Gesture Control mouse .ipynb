```python
# download the libraries by using the following commands
# pip install opencv
# pip install mediapipe
# pip install pyautogui
import cv2
# opencv is used to open and capture the fram via web cam
import  mediapipe as mp
# mediapipe is a powerful tool that provides detecting mechanism
import  time
import pyautogui
# we import puaytogui to perform the various functions of operating system
# creating a dataframe to capture an image from the device 0
cap = cv2.VideoCapture(0)
# priniting all the poins in the hands using mediapipe
mpHands = mp.solutions.hands
hands = mpHands.Hands()
# Displaying each points with some colors
mpDrow = mp.solutions.drawing_utils
# setting up the window size
screen_width, screen_height = pyautogui.size()
index_x=0
Lindex_x=0
index_y=0
id_20y=0
id_16y=0
pTime = 0
cTime =0
while True:
    success,img = cap.read()
    img = cv2.flip(img,1)
    imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    results = hands.process(imgRGB)
    # 2. Get the tip of the index and middle fingers

    #print(results.multi_hand_landmarks)
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            for id,lm in enumerate(handLms.landmark):
                #print(id,lm)
                h,w,c=img.shape
                cx , cy = int(lm.x*w),int(lm.y*h)
                #print(id,cx,cy)
                # index finger  function
                if id==8:
                    cv2.circle(img,(cx,cy), 10 , (255, 255, 255), cv2.FILLED)
                    index_x = screen_width/w*cx
                    index_y = screen_height/h*cy
                    pyautogui.moveTo(index_x,index_y)
                # thumb finger function
                if id==4:
                    cv2.circle(img,(cx,cy), 10 , (255, 0, 255), cv2.FILLED)
                    thumb_x = screen_width/w*cx
                    thumb_y = screen_height/h*cy
                    if thumb_y>thumb_x:
                        print("greater")
                    #print(thumb_y,thumb    _y)
                    #print('outside', abs(int(index_y - thumb_y)))
                    if abs(index_y - thumb_y) < 30:
                        pyautogui.vscroll(-50, x=100, y=100)
                        #print("Scroll")
                # middle finger function
                if id==12:
                    cv2.circle(img,(cx,cy), 10 , (255, 0, 255), cv2.FILLED)
                    middle_x = screen_width/w*cx
                    middle_y = screen_height/h*cy
                    #print(middle_x,middle_y)
                    #print("outside",abs(int(middle_y-thumb_x)))
                    print('outside', abs(int(index_x - middle_x)))
                    #print('test', abs(int(  middle_y-id_9y)))
                    #print('test', abs(int(  middle_y-thumb_y)))
                    if (abs(index_x-middle_x))<50:
                        pyautogui.click()
                        print("Click")
                    if (abs(index_x-middle_x))>250:
                        pyautogui.vscroll(50, x=100, y=100)
                        print("scroll up")
                    if (abs(middle_y-id_9y))<20:
                        print("doubleClick")
                        pyautogui.doubleClick()
                    if abs(middle_y-thumb_y)<30:
                        print("right click")
                        pyautogui.rightClick()
                    if abs(id_20y-thumb_y)<10:
                        pyautogui.press("volumeDown")
                        print("Volume Decrease")
                    if abs(id_16y-thumb_y)<10:
                        pyautogui.press("volumeUP")
                        print("Volume increase")
                # end point of middle finger
                if id==9:
                    cv2.circle(img,(cx,cy), 5 , (0, 255, 255), cv2.FILLED)
                    id_9x = screen_width/w*cx
                    id_9y = screen_height/h*cy
                    #pyautogui.moveTo(index_x,index_y)
                # small finger function
                if id==20:
                    cv2.circle(img,(cx,cy), 10 , (0, 255, 255), cv2.FILLED)
                    id_20x = screen_width/w*cx
                    id_20y = screen_height/h*cy
                    #pyautogui.moveTo(index_x,index_y)
                # ring finger function
                if id==16:
                    cv2.circle(img,(cx,cy), 10 , (0, 255, 255), cv2.FILLED)
                    id_16x = screen_width/w*cx
                    id_16y = screen_height/h*cy
                    #pyautogui.moveTo(index_x,index_y)
                #print('outside', abs(index_y - thumb_y))
                if id==8:
                    cv2.circle(img,(cx,cy), 5 , (0, 255, 255), cv2.FILLED)
                    Lindex_x = screen_width/w*cx
                    Lindex_y = screen_height/h*cy
                    print('match', abs(int(Lindex_x - index_x)))
                    if (abs(Lindex_x-index_x))==0:
                        print("heart")



            mpDrow.draw_landmarks(img,handLms,mpHands.HAND_CONNECTIONS)
    cTime=time.time()
    fps = 1/(cTime-pTime)
    pTime=cTime

    cv2.putText(img,str(int(fps)),(10,70),cv2.FONT_HERSHEY_PLAIN,3,(255,0,255),3)
    cv2.imshow("AI Virtual mouse",img)
    cv2.waitKey(1)
```
